# Stack plot

```python
days=[1,2,3,4,5]
sleeping=[3,4,2,5,7]
eating=[6,5,3,4,2]
working=[6,7,7,8,6]
playing=[5,6,4,3,5]
```

```python
plt.plot([],[], color='m', label= 'Sleeping', linewidth='5')
plt.plot([],[], color='c', label= 'eating', linewidth='5')
plt.plot([],[], color='r', label= 'working', linewidth='5')
plt.plot([],[], color='k', label= 'playing', linewidth='5')
plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])
plt.xlabel(' X label')
plt.ylabel(' Y label')
plt.title(' title ')
plt.legend()
plt.show()
```

# Pie Chart

```python
slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','b', 'r']
```

```python
plt.pie(slices,
       labels=activities,
       shadow=True,
       colors=cols,
       startangle=90,
       explode=(0,0.1,0,0),
        autopct='%1.1f%%')
plt.title('Area of plot')
plt.show()
```
